```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import datasets
df_album=pd.read_csv('/kaggle/input/spotify-dataset-2023/spotify-albums_data_2023.csv')
df_album.head()
```

```python
df_album.describe()
```

```python
#remove columns
df1_album=df_album[['track_name', 'track_id', 'track_number', 'duration_ms', 'album_type',
       'artists', 'total_tracks', 'album_name', 'release_date', 'label',
       'album_popularity', 'album_id', 'artist_id', 
      'duration_sec']]
df1_album=df1_album.sort_values(by='album_popularity',ascending=False)
df1_album.head()
```

```python
#visualise the distribution of songs by their album_popularity score
plt.hist(df1_album['album_popularity'], bins=50, color='skyblue', edgecolor='black')  # Adjust number of bins as needed
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of album_popularity')

plt.show()
```

```python
df1_album['album_popularity'].value_counts()
```

```python
#import spotfify_tracks_data
df_tracks = pd.read_csv('/kaggle/input/spotify-dataset-2023/spotify_tracks_data_2023.csv')

df_tracks.head()
```

```python
#import Spotify_artist_data
df_artists = pd.read_csv('/kaggle/input/spotify-dataset-2023/spotify_artist_data_2023.csv')
df_artists.head()
```

```python
#join with df_album
df_merged = pd.merge(df1_album, df_tracks, left_on='track_id', right_on='id', how='left')
df1_merged = pd.merge(df_merged,df_artists,left_on='artist_id',right_on='id',how='left')
df1_merged['release_date'] = pd.to_datetime(df1_merged['release_date']).dt.strftime('%d-%m-%Y')
df1_merged['release_date'] = pd.to_datetime(df1_merged['release_date'])
df1_merged.head()
```

```python
#remove columns
df2_merged = df1_merged[['track_name','duration_ms', 'album_type',
        'total_tracks', 'album_name', 'release_date', 'label',
       'album_popularity', 'duration_sec',
       'track_popularity', 'explicit', 'name', 'artist_popularity',
       'artist_genres', 'followers','track_id']]
df2_merged= df2_merged.rename(columns={'name': 'artist'})
df2_merged['release_year'] = df2_merged['release_date'].dt.year.astype(str)

df2_merged.head()
```

```python
#top 10 most popular albums releases in 2023
df_top_albums = df2_merged.drop_duplicates(subset=['album_name'])
df_top_albums = df_top_albums[['album_name','artist','album_popularity','release_year','release_date']].sort_values(by='album_popularity',ascending=False)
df_top_albums = df_top_albums[(df_top_albums['release_year'] == '2023.0')].reset_index()
df_top_albums.head(10)
```

```python
#top 10 track in 2023
df_top_tracks = df2_merged.drop_duplicates(subset=['track_name'])
df_top_tracks = df_top_tracks[['track_name','artist','track_popularity','release_year','release_date']].sort_values(by='track_popularity',ascending=False)
df_top_tracks = df_top_tracks[(df_top_tracks['release_year'] == '2023.0')].reset_index()
df_top_tracks.head(10)
```

```python
#import spotify_feautres
df_features = pd.read_csv('/kaggle/input/spotify-dataset-2023/spotify_features_data_2023.csv')
df_features.head()
```

```python
#merge features into df2_merged
df3_merged = pd.merge(df2_merged, df_features, left_on='track_id', right_on='id', how='left')
df3_merged.head()
```

```python
#Drop columns
df1_features_corr = df3_merged[['track_popularity',
       'duration_sec',  'followers', 'danceability', 'energy', 'key', 'loudness', 'mode',
       'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo']]
df1_features_corr.head()
```

```python
#How closely do features correlate to track_popularity
correlation_matrix = df1_features_corr.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()

#The number of followers an artist has is a good indicator of the track's popularity. Other features don't appear to be strongly correlated.
```
